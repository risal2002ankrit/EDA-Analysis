**3.1 Identify Outliers using Z-Scores**

```python
from scipy import stats

# Claculate Z-Score of each Value in dataframe
z_scores_coded = stats.zscore(df_coded.select_dtypes(include=['float64', 'int64']))
print(z_scores_coded)
```

```python
z_scores_en = stats.zscore(df_en.select_dtypes(include=['float64', 'int64']))
print(z_scores_coded)
```

```python
# Identify Outliers
df_outliers_coded = (z_scores_coded > 3).any(axis = 1) | (z_scores_coded < -3).any(axis = 1)
no_of_outliers_coded = df_outliers_coded.sum()
print(f'Number of outliers in df_coded: {no_of_outliers_coded}')
print(df_outliers_coded)
```

```python
df_outliers_en = (z_scores_coded > 3).any(axis = 1) | (z_scores_coded < -3).any(axis = 1)
no_of_outliers_en = df_outliers_en.sum()
print(f'Number of outliers in df_en: {no_of_outliers_en}')
print(df_outliers_en)
```

**3.2 Removing Outliers**

```python
# Removing Row with outliers
df_no_outlier_coded = df_coded[~df_outliers_coded]
df_no_outlier_en = df_en[~df_outliers_en]
```

```python
df_no_outlier_coded.info()
```

```python
print(f"Original number of rows: {df_coded.shape[0]}")
print(f"Number of rows after removing outliers: {df_no_outlier_coded.shape[0]}")
```

```python
print(f"Original number of rows: {df_en.shape[0]}")
print(f"Number of rows after removing outliers: {df_no_outlier_en.shape[0]}")
```
